```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing.sequence import pad_sequences
```

```python
%matplotlib inline
```

```python
def load(chunk):
    # load dataframe from data exploration
    print('Reading JSON chunk', chunk)
    fp = 'saved/urban_sound_classification_data_exp_' + str(chunk) + '.json'
    df = pd.read_json(fp) 
    return df
```

```python
def pad(df, maxlength=96000):
    # pad audio and fft
    print('Padding/Truncating')
    features = df[['Audio', 'Energy', 'Label']]
    
    audio = []
    for item in features.Audio:
        if len(item) < maxlength:
            audio.append(item + (maxlength - len(item)) * [0])
        else:
            audio.append(item[:maxlength])
    features.Audio = audio
    
    fft = []
    halflength = int(maxlength/2)
    for item in features.Energy:
        item = item[:int(len(item)/2)]
        if len(item) < halflength:
            fft.append(item + (halflength - len(item)) * [0])
        else:
            fft.append(item[:halflength])
    features.Energy = fft
    
    return features
```

```python
def save(df, chunk, fp=None):
    # save dataframe from feature engineering
    print('Saving JSON chunk', chunk)
    if not fp:
        fp = 'urban_sound_classification_feature_eng_' + str(chunk) + '.json'
    df.to_json(fp)
    return
```

```python
for i in range(11):
    print("Processing chunk", i)
    df = load(i)
    features = pad(df)
    save(features, i)
```

```python
samples = [features[features.Label==label].iloc[0] for label in set(features.Label)]
print(samples)
```

```python
# set label values for Categorical variable
classes_dict = {item:num for num, item in enumerate(list(set(df.Class)))}
classes_dict
```

```python
window = 1000
rev_classes_dict = {classes_dict[key]:key for key in classes_dict.keys()}
for s in samples:
    tmp = [sum(s.Energy[i:i+window]) for i in range(0, len(s.Energy), window)]
    plt.plot(tmp[10:])
    plt.xlabel("Windowed Frequency")
    plt.ylabel("Sum of Energy")
    plt.title(rev_classes_dict[s.Label])
    plt.show()
```

```python
window = 2000
rev_classes_dict = {classes_dict[key]:key for key in classes_dict.keys()}
for s in samples:
    tmp = [np.mean(s.Audio[i:i+window]) for i in range(0, len(s.Audio), window)]
    plt.plot(tmp[10:])
    plt.xlabel("Windowed Time")
    plt.ylabel("Mean of Frequency")
    plt.title(rev_classes_dict[s.Label])
    plt.show()
```

```python
def window(df, size=1000):
    print('Windowing Audio')
    df['Audio_Windowed'] = [[np.mean(s[i:i+size]) for i in range(0, len(s), size)] for s in df.Audio]
    print('Windowing Energy')
    size = int(size/2)
    df['Energy_Windowed'] = [[sum(s[i:i+size]) for i in range(0, len(s), size)] for s in df.Energy]
    df = df[['Audio_Windowed', 'Energy_Windowed', 'Label']]
    return df
```

```python
for i in range(11):
    print("Processing chunk", i)
    df = load(i)
    features = pad(df)
    features_windowed = window(features, 2000)
    save(features_windowed, i)
```
